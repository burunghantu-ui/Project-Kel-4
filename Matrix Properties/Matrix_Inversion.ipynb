**SOAL NO 1**

```python
import numpy as np

X = np.array([[1, 2],
              [3, 4]])

print("Matrix X:")
print(X)

Xinv = np.linalg.inv(X)

print("\nInvers Matrix X (Xinv):")
print(Xinv)

hasil = np.dot(X, Xinv)

print("\nHasil perkalian X dengan Xinv:")
print(hasil)
```

**SOAL NO 2**

Menggunakan numpy

```python
import numpy as np

A = np.array([[4, 2],
              [-5, -3]])

B = np.array([4, -7])

X = np.linalg.solve(A, B)

print("Solusi (b, c) =", X)
```

Menggunakan tensorflow

```python
import tensorflow as tf

A = tf.constant([[4.0, 2.0],
                 [-5.0, -3.0]])

B = tf.constant([[4.0],
                 [-7.0]])

X = tf.linalg.solve(A, B)

print("Solusi (b, c) =", X.numpy())
```

Menggunakan torch

```python
import torch

A = torch.tensor([[4.0, 2.0],
                  [-5.0, -3.0]])

B = torch.tensor([[4.0],
                  [-7.0]])

X = torch.linalg.solve(A, B)

print("Solusi (b, c) =", X)
```
