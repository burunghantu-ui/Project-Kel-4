Soal No 1.

```python
import numpy as np
# buat matrix A dengan ukuran 2x3 dan vector b berukuran 3x1 menggunakan numpy
A = np.array([[12, 2, 7], [9, 5, 3]])
b = np.array([[2], [8], [19]])

print("Matrix A:")
print(A)

print("\nVector b:")
print(b)
```

```python
import torch
# buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
A_pt = torch.tensor([[12, 2, 7], [9, 5, 3]])
b_pt = torch.tensor([[2], [8], [19]])

print("Matrix A_pt:")
print(A_pt)

print("\nVector b_pt:")
print(b_pt)
```

```python
import tensorflow as tf
# buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
A_tf = tf.Variable([[12, 2, 7], [9, 5, 3]])
b_tf = tf.Variable([[2], [8], [19]])

print("Matrix A_tf:")
print(A_tf)

print("\nVector b_tf:")
print(b_tf)
```

```python
# kalikan matrix A_pt dengan vector b_pt
A_ptb_pt = torch.matmul(A_pt, b_pt)

print("Hasil kali A_pt dengan b_pt adalah:")
print(A_ptb_pt)
```

Soal No 4

```python
# buat matrix B dengan ukuran 3x3 menggunakan numpy
B = np.array([[3, 1, 4], [9, 8, 6], [7, 2, 10]])

print("Matrix B:")
print(B)
```

```python
# buat matrix B_pt dengan ukuran 3x3 menggunakan pytorch
B_pt = torch.tensor([[3, 1, 4], [9, 8, 6], [7, 2, 10]])

print("Matrix B_pt:")
print(B_pt)
```

```python
# buat matrix B_tf dengan ukuran 3x3 menggunakan tensorflow
B_tf = tf.Variable([[3, 1, 4], [9, 8, 6], [7, 2, 10]])

print("Matrix B_tf:")
print(B_tf)
```

Soal No 5

```python
# kalikan matrix A dengan matrix B
AB = np.dot(A, B)

print("Hasil kali Matrix A dengan B adalah:")
print(AB)
```

```python
# kalikan matrix A_pt dengan matrix B_pt
A_ptB_pt = torch.matmul(A_pt, B_pt)

print("Hasil kali A_pt dengan B_pt adalah:")
print(A_ptB_pt)
```

```python
# kalikan matrix A_tf dengan matrix B_tf
A_tfB_tf = tf.matmul(A_tf, B_tf)

print("Hasil kali A_tf dengan B_tf adalah:")
print(A_tfB_tf)
```
