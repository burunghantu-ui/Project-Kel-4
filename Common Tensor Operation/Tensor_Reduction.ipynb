**Soal No. 1**

```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy

import numpy as np

X = np.array([[1, 2, 3], [4, 5, 6]])

print(f"matrix X: \n{X}")
```

```python
# hitung berapa hasil penjumlahan elemen matrix X dengan fungsi sum

print(f"hasil penjumlahan seluruh elemen matrix X adalah {X.sum()}")
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow

import tensorflow as tf

X_tf = tf.Variable([[7, 8, 9], [10, 11, 12]])

print(f"matrix X_tf: \n{X_tf}")
```

**Soal No. 2**

```python
# hitung hasil penjumlahan seluruh elemen matrix X

print(f"hasil penjumlahan seluruh elemen matrix X adalah {X.sum()}")
```

```python
# hitung hasil penjumlahan seluruh elemen matrix X_pt

import torch

X_pt = torch.tensor([[13, 14, 15], [16, 17, 18]])

print(f"hasil penjumlahan seluruh elemen matrix X_pt adalah {torch.sum(X_pt)}")
```

```python
# hitung hasil penjumlahan seluruh elemen matrix X_tf

print(f"hasil penjumlahan seluruh elemen matrix X_tf adalah {tf.reduce_sum(X_tf)}")
```

**Soal No. 3**

```python
# hitung hasil penjumlahan matrix X berdasarkan row nya

print(f"hasil penjumlahan seluruh kolom dari matrix X adalah, matrix X: {X.sum(axis=1)}")
```

```python
# hitung hasil penjumlahan matrix X_pt berdasarkan row nya

print(f"hasil penjumlahan seluruh kolom dari matrix X_pt adalah, matrix X_pt: {torch.sum(X_pt, axis=1)}")
```

```python
# hitung hasil penjumlahan matrix X_tf berdasarkan row nya

print(f"hasil penjumlahan seluruh kolom dari matrix X_ft adalah, matrix X_ft: {tf.reduce_sum(X_tf, axis=1)}")
```

**Soal No. 4**

```python
# hitung hasil penjumlahan matrix X berdasarkan kolom nya

print("4. Penjumlahan kolom X     :", np.sum(X, axis=0))
```

```python
# hitung hasil penjumlahan matrix X_pt berdasarkan kolom nya

print("4. Penjumlahan kolom X_pt  :", torch.sum(X_pt, dim=0).numpy())
```

```python
# hitung hasil penjumlahan matrix X_tf berdasarkan kolom nya

print("4. Penjumlahan kolom X_tf  :", tf.reduce_sum(X_tf, axis=0).numpy())
```

**Soal No. 5**

```python
# hitung nilai minimal dari matrix X, X_pt dan X_tf

print("5. Nilai minimal  -> X:", np.min(X), "| X_pt:", torch.min(X_pt).item(), "| X_tf:", tf.reduce_min(X_tf).numpy())
```

```python
# hitung nilai maksimal dari matrix X, X_pt dan X_tf

print("5. Nilai maksimal -> X:", np.max(X), "| X_pt:", torch.max(X_pt).item(), "| X_tf:", tf.reduce_max(X_tf).numpy())
```

```python
# hitung nilai rata-rata dari matrix X, X_pt dan X_tf

print("5. Rata-rata      -> X:", np.mean(X), "| X_pt:", torch.mean(X_pt.float()).item(), "| X_tf:", tf.reduce_mean(X_tf).numpy())
```

**Soal No. 6**

```python
# # hitung hasil perkalian seluruh elemen matrix X

print("6. Perkalian elemen X      :", np.prod(X))
```

```python
# hitung hasil perkalian seluruh elemen matrix X_pt

print("6. Perkalian elemen X_pt   :", torch.prod(X_pt).item())
```

```python
# hitung hasil perkalian seluruh elemen matrix X_tf

print("6. Perkalian elemen X_tf   :", tf.reduce_prod(X_tf).numpy())
```
