```python
# buat matrix X dengan ukuran 2x3 menggunakan numpy

import numpy as np

X = np.array([[1, 2, 3], [4, 5, 6]])

print(f"matrix X:")
print(X)
```

```python
# buat matrix X_tf dengan ukuran 2x3 menggunakan tensorflow

import tensorflow as tf

X_tf = tf.constant([[13, 14, 15], [16, 17, 18]])

print(f"matrix X_ft:")
print(X_tf)
```

**Soal No. 2**

```python
# buat matrix Y dengan ukuran 2x3 menggunakan numpy

import numpy as np

Y = np.array([[1, 2, 3], [4, 5, 6]])

print(f"matrix Y: \n{Y}")
```

```python
# buat matrix Y_pt dengan ukuran 2x3 menggunakan pytorch

import torch

Y_pt = torch.tensor([[7, 8, 9], [10, 11, 12]])

print(f"matrix Y_pt menggunakan pytorch: \n{Y_pt}")
```

```python
# buat matrix Y_tf dengan ukuran 2x3 menggunakan tensorflow

import tensorflow as tf

Y_tf = tf.Variable([[13, 14, 15], [16, 17, 18]])

print(f"matrix Y_tf menggunakan tensorflow: \n{Y_tf}")
```

```python
# kalikan matrix X dengan Y

print(f"hasil perkalian matrix X dengan matriks Y adalah \n{X * Y}")
```

```python
#	kalikan matrix X_pt dengan Y_pt

print(f"hasil perkalian matrix X_pt dengan matriks Y_pt adalah \n{X_pt * Y_pt}")
```

```python
# kalikan matrix X_tf dengan Y_tf

print(f"hasil perkalian matrix X_tf dengan matriks Y_tf adalah \n{X_tf * Y_tf}")
```
