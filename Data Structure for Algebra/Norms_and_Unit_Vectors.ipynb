**SOAL NO.1**

```
import numpy as np

x = np.array([[1], [2], [3]])

norm_x = np.linalg.norm(x)

print("Vector x:\n", x)
print("L2 norm x:", norm_x)
```

**SOAL NO.2**

```
v = np.array([3, 4])

norm_v = np.linalg.norm(v)

unit_v = v / norm_v

norm_unit_v = np.linalg.norm(unit_v)

print("Vector v:", v)
print("L2 norm v:", norm_v)
print("Unit vector v:", unit_v)
print("Panjang unit vector:", norm_unit_v)
```
