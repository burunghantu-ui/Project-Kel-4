**SOAL NO.1**

```
# buat variabel skalar x dan simpan suatu nilai skalar
x = 10
print("Nilai x:", x)

# tampilkan tipe data dari x
print("Tipe data x:", type(x))

# buat variabel skalar y dan simpan suatu nilai skalar
y = 5
print("Nilai y:", y)

# tampilkan tipe data dari y
print("Tipe data y:", type(y))

# jumlahkan x+y
hasil_xy = x + y
print("Hasil x + y:", hasil_xy)

# buat variabel bernama x_float dan simpan suatu nilai desimal
x_float = 3.5
print("Nilai x_float:", x_float)

#jumlahkan x_float dengan y
hasil_float = x_float + y
print("Hasil x_float + y:", hasil_float)

# tampilkan tipe data dari x_float
print("Tipe data x_float:", type(x_float))
```

**SOAL NO.2**

```
# lakukan tahapan pada soal 1 dengan menggunakan tensorflow

import tensorflow as tf

x = tf.constant(10)
print("\nNilai x:", x)
print("Tipe data x:", x.dtype)

y = tf.constant(5)
print("\nNilai y:", y)
print("Tipe data y:", y.dtype)

hasil_xy = x + y
print("\nHasil x + y:", hasil_xy)

x_float = tf.constant(3.5)
print("\nNilai x_float:", x_float)

# rumus "tf_sum = tf.add(x_float, y_tf)" tidak bisa digunakan"
hasil_float = x_float + tf.cast(y, tf.float32)
print("\nHasil x_float + y:", hasil_float)

print("Tipe data x_float:", x_float.dtype)
```

**SOAL NO.3**

```
import torch

x = torch.tensor(10)
print("\nNilai x:", x)
print("Tipe data x:", x.dtype)

y = torch.tensor(5)
print("\nNilai y:", y)
print("Tipe data y:", y.dtype)

hasil_xy = x + y
print("\nHasil x + y:", hasil_xy)

x_float = torch.tensor(3.5)
print("\nNilai x_float:", x_float)

hasil_float = x_float + y
print("\nHasil x_float + y:", hasil_float)

print("\nTipe data x_float:", x_float.dtype)
```
