```python
import numpy as np
```

```python
# 1. Buat matriks X berukuran 2x3
X = np.array([[10, 20, 30], [40, 50, 60]])
print("Matriks X:\n", X)
```

```python
# Tampilkan ukuran dari matriks X
print("Ukuran matriks X:", X.shape)
```

```python
# Tampilkan seluruh elemen dari kolom pertama (index 0)
print("Kolom pertama:", X[:, 0])
```

```python
# Tampilkan seluruh elemen dari baris kedua (index 1)
print("Baris kedua:", X[1, :])
```

2. Ulangi soal no 1 menggunakan pytorch

```python
import torch
```

```python
# Membuat matrix 2x3
X = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print("Matriks X:\n", X)
```

```python
# Menampilkan ukuran matrix
print("Ukuran matrix X:", X.shape)
```

```python
# Menampilkan kolom pertama
print("Kolom pertama:", X[:, 0])
```

```python
# Menampilkan baris kedua
print("Baris kedua:", X[1, :])
```
